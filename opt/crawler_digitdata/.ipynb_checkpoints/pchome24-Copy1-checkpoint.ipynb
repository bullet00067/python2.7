```python
import requests
res = []
for page in range(1,3):
    ress = requests.get("https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=ssd&page="+str(page)+"&sort=sale/dc")
    res.append(ress)
```

```python
res
```

```python
flattened_list = []
for x in res:
    for y in x:
        flattened_list.append(y)
srcinput = ""
for idx in range(1,len(flattened_list)):
    srcinput+=flattened_list[idx].decode('utf-8')
srcinput
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(srcinput, 'lxml')
soup
```

```python
import json
data = json.load(srcinput)
data.keys()
```

```python
data['prods']
```
